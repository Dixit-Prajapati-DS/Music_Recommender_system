```python
min_max_scaler=['energy','key','mode','speechiness','liveness','valence','time_signature']
standard_scaler=['duration_ms','danceability','loudness','acousticness','instrumentalness','tempo']
text =['tag']
```

```python
from nltk.stem.porter import PorterStemmer

po= PorterStemmer()
def stem(x):
    y=[]
    for i in x.split():
        y.append(po.stem(i))
    return " ".join(y)

song['tag']=song['tag'].apply(stem)
    
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
p1= ColumnTransformer([
    ('min_max_scaled',MinMaxScaler(),min_max_scaler),
    ('st_scaled',StandardScaler(),standard_scaler),
    ('Vectors',CountVectorizer(stop_words='english',max_features=1000),'tag')
],remainder='drop')
```

```python
vectorize= p1.fit_transform(song)
```

```python
# d= p1.get_feature_names_out()

# for i in d:
#     print(i)
```

```python
# Recommend_index= 3748
song_input = song[song.index==3748]
```

```python
s_input= p1.transform(song_input)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vectorize,s_input)
```

```python
sorted(list(enumerate(similarity.tolist())),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def Recommend(name):
    index = song[song["name"]==name].index[0]
    song_input=song[song["name"]==name]
    s_input= p1.transform(song_input)

    Similarity_score= cosine_similarity(vectorize,s_input)
    Result = sorted(list(enumerate(Similarity_score.tolist())),reverse=True,key=lambda x:x[1])[1:6]
    movie_name=[]
    track_id=[]
    for i in Result:
        movie_name.append(song.iloc[i[0],1])
        track_id.append(song.iloc[i[0],0])
    return movie_name,track_id
```

```python
name,track_id = Recommend('Death Threat')
```
